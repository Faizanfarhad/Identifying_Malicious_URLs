```python
import torch as pt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
```

### Data visualization

```python
df = pd.read_csv('/home/faizan/Malicious_URLs/src/Dataset/malicious_phish.csv')

df.info()
```

#### Info

```python
plt.figure(figsize=(10,6))

sns.histplot(data=df,x='type',color='green',palette='magma',hue='type')
```

#### Malware Datapoints

```python

plt.figure(figsize=(10,6))
plt.ylim(32000,50000)
sns.histplot(data=df,x='type',color='green',palette='rocket',hue='type')
```

#### Phishing and Defacement datapoints

```python
plt.figure(figsize=(10,6))
plt.ylim(92000,100000)
sns.histplot(data=df,x='type',color='green',palette='rocket_r',hue='type')
```

#### Benign or Not Harmful URLs Datapoints
